OverTime vs. Marital Status

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='MaritalStatus', hue='OverTime', data=df)
plt.title('MaritalStatus vs. OverTime')
plt.xlabel('MaritalStatus')
plt.ylabel('Count')
plt.legend(title='OverTime')
plt.show()
```
